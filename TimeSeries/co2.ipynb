```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

### 1. Wgraj zbiór danych CO2

```python
df = pd.read_csv('Co2 dataset.csv')
df.head()
```

### 2. Stwórz wykres pamiętając o podpisaniu osi

```python
fig, ax = plt.subplots(1,1,figsize=(9,5))

sns.lineplot(df, x='Decimal Date', y='Carbon Dioxide (ppm)',ax = ax)
ax.set(xlabel='Years', ylabel='CO2 concentration', title='CO2 in different years')

plt.show()
```

### 3. Określ sezonowość i wylicz krzywą trendu. Przykładowo dla danych o częstotliwości miesięcznej oraz rocznej sezonowości zjawiska potrzebne będzie stworzenie okna w rozmiarze 12. Jeśli okresowość naszego zjawiska jest liczbą nieparzystą to średnią kroczącą do wyliczenia trendu możemy obliczyć dla okna 1xT, natomiast gdy jest liczbą parzystą używamy okna 2xT. Wykonujemy ten krok by pozbyć się fluktuacji sezonowych.

```python
window_len = 12
df["trend"] =(df['Carbon Dioxide (ppm)'].rolling(window=window_len).mean().rolling(window=2).mean().shift(-window_len // 2))
```

### 4. Wykonaj wizualizację sygnału wejściowego oraz trendu na jednym wykresie

```python
fig, ax = plt.subplots(1,1,figsize=(9,5))

sns.lineplot(df, x='Decimal Date', y='Carbon Dioxide (ppm)',ax = ax, label='Real')
sns.lineplot(df, x='Decimal Date', y='trend', ax=ax, label='Trend')
ax.set(xlabel='Years', ylabel='CO2 concentration', title='CO2 in different years')

plt.show()
```

### 5. Odejmij od sygnału wejściowego wyliczony trend i wykonaj na jednym wykresie wykres wszystkich trzech składowych w celu sprawdzenia. Skomentuj KRÓTKO wynik

```python
df["Input_detrended"] = df["Carbon Dioxide (ppm)"] - df["trend"]
df.head()
```

```python
fig, ax = plt.subplots(1,1,figsize=(9,5))

sns.lineplot(df, x='Decimal Date', y='Carbon Dioxide (ppm)',ax = ax, label='Real')
sns.lineplot(df, x='Decimal Date', y='trend', ax=ax, label='Trend')
sns.lineplot(df, x='Decimal Date', y='Input_detrended', ax=ax, label='Detrended real')
ax.set(xlabel='Years', ylabel='CO2 concentration', title='CO2 in different years')

plt.show()
```

Wynikowa krzywa jest 'prosta' (w sensie kierunku pionowego) co oznacza, że udało nam się pozbyć trendu i posiadamy jedynie dane zawierające sezonowość i rezydua

### 6. Wylicz średnia szeregu czasowego po odfiltrowaniu trendu dla danej częstotliwości i skomentuj wynik.

```python
seasonality = df.groupby("Month").mean()["Input_detrended"]
seasonality.name = "seasonality"
fig, ax = plt.subplots(figsize=[12, 5])
seasonality.plot(y="seasonality", ax=ax, marker="o", color="red", markersize=18)
ax.set_xlabel("Month", fontsize=18)
ax.set_ylabel("Seasonality", fontsize=18)
ax.set_title("Time series mean after filtering out the trend for each month", fontsize=18)
plt.tight_layout()
plt.show()
```

### 7. Powtórz pattern sezonowości wyliczony w kroku 6 i oblicz rezydua. Przykład:

```python
df = df.merge(right=seasonality, left_on="Month", right_index=True)
df = df.sort_index()
df.head()
```

```python
df["residual"] = df["Carbon Dioxide (ppm)"] - df["trend"] - df["seasonality"]
df.head()
```

### 8. Wykonaj wykres dla rezyduów 

```python
fig, ax = plt.subplots(1,1,figsize=(9,5))

sns.lineplot(df, x='Decimal Date', y='residual',ax = ax)
ax.set(xlabel='Date', ylabel='Residuals', title='Residuals')

plt.show()
```

### 9. Wykonaj zbiorczy wykres i krótki zinterpretuj wyniki. Przykład:

```python
fig, ax = plt.subplots(nrows=4, figsize=[12, 12], sharex=True)

df["Carbon Dioxide (ppm)"].plot(ax=ax[0], legend="Real")
ax[0].set_ylabel("CO2")

df["trend"].plot(ax=ax[1], legend="trend")
ax[1].set_ylabel("Trend")

df["seasonality"].plot(ax=ax[2], legend="seasonality")
ax[2].set_ylabel("Seasonal component")

df["residual"].plot(ax=ax[3], marker=".", legend="residual", ls="None")
ax[3].set_ylabel("Residuals")

ax[0].set_title("CO2 analysis")
ax[3].set_xlabel("Date")

plt.tight_layout()
```
