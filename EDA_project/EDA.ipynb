### Podsumowanie

Po wszystkich przeprowadzonych analizach postanowiono (i wykonano bądź zostanie to wykonane w poniższych komórkach):
- usunięciu duplikatów jeśli różnią się tylko wartościami w kolumnie n_type_stations (nie jest to informacja użyteczna z punktu widzenia analizy więc brak tej kolumny nie przeszkadza) - zrobione
- dodaniu kolumny country_city do analiz na poziomie miasta - zrobione
- analizie wartości tylko, jeśli jej roczny coverage przekracza 70% (wartość arbitralna) i odrzuceniu pozostałych wartości, jeśli warunek ten nie jest spełniony. Jednocześnie zadecydowano o nie uzupełnianiu brakujących wartości z powodu zbyt skomplikowanego zjawiska jakim jest zanieczyszczenie oraz dużych różnic czasowych i przestrzennych między poszczególnymi punktami.
- usunięciu wartości różniących się o więcej niż 3 odchylenia standardowe od średniej w górę (w dół taki warunek nie musi być sprawdzany ponieważ zanieczyszczenie nie może być ujemne)
- usunięciu zbędnych kolumn

```python
data_final = data_correct_cities.copy()

pm10_bool = (data_final.pm10_coverage > 70) & (data_final.pm10 < data_final.pm10.mean() +  data_final.pm10.std()*3)
pm25_bool = (data_final.pm25_coverage > 70) & (data_final.pm25 < data_final.pm25.mean() + data_final.pm25.std()*3)
no2_bool = (data_final.no2_coverage > 70) & (data_final.no2 < data_final.no2.mean() + data_final.no2.std()*3)

pm10_new = np.where(pm10_bool == True, data_final.pm10, None)
pm25_new = np.where(pm25_bool == True, data_final.pm25, None)
no2_new = np.where(no2_bool == True, data_final.no2, None)

data_final['pm10_new'] = pm10_new
data_final['pm25_new'] = pm25_new
data_final['no2_new'] = no2_new

data_final = data_final.dropna(subset=['pm10_new', 'pm25_new', 'no2_new'], how='all', axis=0)
data_final = data_final[['who_region', 'iso3', 'country_city', 'country', 'city','year', 'pm10_new', 'pm25_new', 'no2_new', 'pm10_coverage', 'pm25_new', 'no2_coverage']]

print(len(data_final))
```

## Mapa

```python
locations = pd.read_csv('worldcities.csv')
unique_places = data_final[['city', 'iso3']].drop_duplicates().reset_index(drop=True)
unique_places_locs = unique_places.merge(locations, how='left', left_on=['city', 'iso3'], right_on=['city_ascii', 'iso3'])[['city_x','iso3','lat', 'lng']]
print(unique_places_locs.isna().sum()/len(unique_places_locs))
```

Używam zbioru danych posiadających współrzędne dla największych miast na świecie. Niestety, brakuje danych dla ok. 50% miast dla których mamy dane - nie powinno to jednak stanowiść problemu, ponieważ największe miasta (w szczególności stolice) są uwzględnione w tym zbiorze, a mapa nie będzie miała na tyle dużej rozdzielczości by było to rozpoznawalne

```python
grouped_data = data_final.groupby(['who_region','year','iso3', 'city'])[['pm10_new', 'pm25_new', 'no2_new']].mean().reset_index()
grouped_locs = grouped_data.merge(unique_places_locs, left_on=['iso3', 'city'], right_on=['iso3','city_x'], how='inner')

grouped_locs
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))

def AnimationFunction(frame):
    years = list(range(2010,2022))
    temp = grouped_locs[grouped_locs.year == years[frame]]
    sns.scatterplot(temp, x='lng', y='lat', hue='pm10_new',  palette='rainbow', size=3)
    ax.get_legend().remove()

ax.set(aspect='equal', xlim=[-180,180], ylim=[-90, 90], title='Średnie wartości PM10 na przestrzeni lat')

anim_created = FuncAnimation(fig, AnimationFunction, frames=10, interval=150)


video = anim_created.to_html5_video()
html = display.HTML(video)
display.display(html) 

plt.close()
```

Na powyższej animacji można zauważyć różnice w średnich wartościach poszczególnych zanieczyszczeń na przestrzeni lat. Wnioski:
- W Europie widać stopniową poprawę jakości powietrza
- Poza Europą - można zauważyć zwiększenie ilości czujników
- Niektóre kraje posiadają zauważalnie gorszą jakość powietrza - Meksyk, Chiny, Indie, kraje Europy południowo-wschodniej

## Związki przyczynowo-skutkowe
- Wartości zanieczyszczeń powietrza są ze sobą silnie skorelowane (nie jest to zbyt zadziwiające)
- Wartości coverage są lekko ujemnie skorelowane - można uznać to za mało istotną informację, ale warto zauważyć, że przy niższym coverage średnie wartości zanieczyszczeń są wyższe - co może powodować obciążenie danych
- W krajach bogatszych, lepiej rozwiniętych znajduje się więcej czujników, a także jakość powietrza stopniowo się polepsza

## Zastosowanie w ML
- Ze względu na dużą granularność a także znaczące braki danych (ciągłości w danych regionach) możliwości użycia tego zbioru danych są ograniczone
- Mogą posłużyć do prognoz jakości powietrza w danych krajach jako ogólne modelowanie trendu (przyszłych lat)
- Zastosowanie byłoby jeszcze lepsze, gdybyśmy do tych danych dopasowali informacje nt. populacji czy np. GDP - pozwoliłoby to modelować zmiany jakości powietrza na przestrzeni lat z uwzględnieniem czynników demograficznych i gospodarczych

4. Parametr 'coverage' jest bardzo istotny, wręcz kluczowy dla analiz, bo oznacza procent czasu w roku z którego zebrane są pomiary. Jeśli jest on zbyt mały, lub nie znamy jego wartości to wartości zanieczyszczeń mogą być silnie przeszacowane (możliwy powody to: włączanie czujników tylko w dzień, kiedy są o wiele wyższe wartości zanieczyszczeń, zbieranie pomiarów kiedy podejrzewana jest gorsza jakość powietrza w wyniku pory roku, działanie czujników na zasadzie fotoradarów - pomiar jest zbierany jeśli przekracza konkretną wartość/normę). Uwzględnienie tego parametru pozwala znacznie ograniczyć ilość wartości błędnych, zmniejszyć ilość outlierów i ogólnie poprawić jakość analiz
