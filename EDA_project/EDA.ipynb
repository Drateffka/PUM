### Podsumowanie

Po wszystkich przeprowadzonych analizach postanowiono (i wykonano bd zostanie to wykonane w poniszych komrkach):
- usuniciu duplikatw jeli rni si tylko wartociami w kolumnie n_type_stations (nie jest to informacja uyteczna z punktu widzenia analizy wic brak tej kolumny nie przeszkadza) - zrobione
- dodaniu kolumny country_city do analiz na poziomie miasta - zrobione
- analizie wartoci tylko, jeli jej roczny coverage przekracza 70% (warto arbitralna) i odrzuceniu pozostaych wartoci, jeli warunek ten nie jest speniony. Jednoczenie zadecydowano o nie uzupenianiu brakujcych wartoci z powodu zbyt skomplikowanego zjawiska jakim jest zanieczyszczenie oraz duych rnic czasowych i przestrzennych midzy poszczeglnymi punktami.
- usuniciu wartoci rnicych si o wicej ni 3 odchylenia standardowe od redniej w gr (w d taki warunek nie musi by sprawdzany poniewa zanieczyszczenie nie moe by ujemne)
- usuniciu zbdnych kolumn

```python
data_final = data_correct_cities.copy()

pm10_bool = (data_final.pm10_coverage > 70) & (data_final.pm10 < data_final.pm10.mean() +  data_final.pm10.std()*3)
pm25_bool = (data_final.pm25_coverage > 70) & (data_final.pm25 < data_final.pm25.mean() + data_final.pm25.std()*3)
no2_bool = (data_final.no2_coverage > 70) & (data_final.no2 < data_final.no2.mean() + data_final.no2.std()*3)

pm10_new = np.where(pm10_bool == True, data_final.pm10, None)
pm25_new = np.where(pm25_bool == True, data_final.pm25, None)
no2_new = np.where(no2_bool == True, data_final.no2, None)

data_final['pm10_new'] = pm10_new
data_final['pm25_new'] = pm25_new
data_final['no2_new'] = no2_new

data_final = data_final.dropna(subset=['pm10_new', 'pm25_new', 'no2_new'], how='all', axis=0)
data_final = data_final[['who_region', 'iso3', 'country_city', 'country', 'city','year', 'pm10_new', 'pm25_new', 'no2_new', 'pm10_coverage', 'pm25_new', 'no2_coverage']]

print(len(data_final))
```

## Mapa

```python
locations = pd.read_csv('worldcities.csv')
unique_places = data_final[['city', 'iso3']].drop_duplicates().reset_index(drop=True)
unique_places_locs = unique_places.merge(locations, how='left', left_on=['city', 'iso3'], right_on=['city_ascii', 'iso3'])[['city_x','iso3','lat', 'lng']]
print(unique_places_locs.isna().sum()/len(unique_places_locs))
```

Uywam zbioru danych posiadajcych wsprzdne dla najwikszych miast na wiecie. Niestety, brakuje danych dla ok. 50% miast dla ktrych mamy dane - nie powinno to jednak stanowi problemu, poniewa najwiksze miasta (w szczeglnoci stolice) s uwzgldnione w tym zbiorze, a mapa nie bdzie miaa na tyle duej rozdzielczoci by byo to rozpoznawalne

```python
grouped_data = data_final.groupby(['who_region','year','iso3', 'city'])[['pm10_new', 'pm25_new', 'no2_new']].mean().reset_index()
grouped_locs = grouped_data.merge(unique_places_locs, left_on=['iso3', 'city'], right_on=['iso3','city_x'], how='inner')

grouped_locs
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))

def AnimationFunction(frame):
    years = list(range(2010,2022))
    temp = grouped_locs[grouped_locs.year == years[frame]]
    sns.scatterplot(temp, x='lng', y='lat', hue='pm10_new',  palette='rainbow', size=3)
    ax.get_legend().remove()

ax.set(aspect='equal', xlim=[-180,180], ylim=[-90, 90], title='rednie wartoci PM10 na przestrzeni lat')

anim_created = FuncAnimation(fig, AnimationFunction, frames=10, interval=150)


video = anim_created.to_html5_video()
html = display.HTML(video)
display.display(html) 

plt.close()
```

Na powyszej animacji mona zauway rnice w rednich wartociach poszczeglnych zanieczyszcze na przestrzeni lat. Wnioski:
- W Europie wida stopniow popraw jakoci powietrza
- Poza Europ - mona zauway zwikszenie iloci czujnikw
- Niektre kraje posiadaj zauwaalnie gorsz jako powietrza - Meksyk, Chiny, Indie, kraje Europy poudniowo-wschodniej

## Zwizki przyczynowo-skutkowe
- Wartoci zanieczyszcze powietrza s ze sob silnie skorelowane (nie jest to zbyt zadziwiajce)
- Wartoci coverage s lekko ujemnie skorelowane - mona uzna to za mao istotn informacj, ale warto zauway, e przy niszym coverage rednie wartoci zanieczyszcze s wysze - co moe powodowa obcienie danych
- W krajach bogatszych, lepiej rozwinitych znajduje si wicej czujnikw, a take jako powietrza stopniowo si polepsza

## Zastosowanie w ML
- Ze wzgldu na du granularno a take znaczce braki danych (cigoci w danych regionach) moliwoci uycia tego zbioru danych s ograniczone
- Mog posuy do prognoz jakoci powietrza w danych krajach jako oglne modelowanie trendu (przyszych lat)
- Zastosowanie byoby jeszcze lepsze, gdybymy do tych danych dopasowali informacje nt. populacji czy np. GDP - pozwolioby to modelowa zmiany jakoci powietrza na przestrzeni lat z uwzgldnieniem czynnikw demograficznych i gospodarczych

## Podsumowanie

Najwaniejsze cechy danych, wnioski i spostrzeenia:
1. Dane posiadamy w praktyce od 2010 roku (wczeniej dane s tylko dla jednego kraju i 3 miast), w szczytowym okresie dla ok. 4000 miast. Liczba miast i krajw z ktrych posiadamy pomiary stale rosa a do 2020 roku kiedy zmalaa (najprawdopodobniej jako efekt uboczny wybuchu pandemii, byy inne priorytety, lub poniewa dane nie zostay jeszcze dostarczone z wszystkich krajw).

Jeli jednak posiadamy dane o dwch, a nawet 3 wartociach to s one ze sob silnie pozytywnie skorelowane (im wiksza jedna warto tym wiksza druga)
4. Parametr 'coverage' jest bardzo istotny, wrcz kluczowy dla analiz, bo oznacza procent czasu w roku z ktrego zebrane s pomiary.
